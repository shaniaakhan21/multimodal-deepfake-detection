```python
import os

# Paths — ✅ Set these correctly
real_videos = "data/faceforensics/original_sequences"
fake_videos = "data/faceforensics/manipulated_sequences/DeepFakeDetection/c23/videos"
real_audio_out = "data/audio_wav/real"
fake_audio_out = "data/audio_wav/fake"

# ✅ Ensure output folders exist
os.makedirs(real_audio_out, exist_ok=True)
os.makedirs(fake_audio_out, exist_ok=True)

# ✅ Function to extract audio using ffmpeg CLI
def extract_audio_from_folder(video_folder, save_folder):
    video_files = [f for f in os.listdir(video_folder) if f.endswith(".mp4")]
    count = 0

    for video in video_files:
        video_path = os.path.join(video_folder, video)
        audio_path = os.path.join(save_folder, video.replace(".mp4", ".wav"))

        # Run ffmpeg shell command silently
        command = f'ffmpeg -loglevel error -i "{video_path}" -vn -acodec pcm_s16le -ar 16000 -ac 1 "{audio_path}" -y'
        os.system(command)
        count += 1

    return count

# ✅ Run for both real and fake
real_count = extract_audio_from_folder(real_videos, real_audio_out)
fake_count = extract_audio_from_folder(fake_videos, fake_audio_out)

print(f"✅ Extracted audio from {real_count} real videos and {fake_count} fake videos.")
```
